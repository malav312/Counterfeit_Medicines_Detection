```python
import pandas as pd
import numpy as np
from scipy import stats

# read in the input csv file
input_data = pd.read_csv('scores.csv')

# create an empty output dataframe with the columns 'Medical Name' and 'Upper Threshold'
output_data = pd.DataFrame(columns=['Medical Name', 'Upper Threshold'])

# iterate over each medical name in the first column of the input dataframe
for medical_name in input_data.iloc[:,0].unique():

    rows_for_medical_name = input_data.loc[input_data.iloc[:,0] == medical_name]

    values_for_medical_name = rows_for_medical_name.iloc[:,1:]
    print(medical_name)
    
#     # select the rows in the input dataframe corresponding to the current medical name
#     rows_for_medical_name = input_data.loc[input_data.iloc[:,0] == medical_name]
    
#     # extract the values for the current medical name from the selected rows
#     values_for_medical_name = rows_for_medical_name.iloc[:,1:]
    
    # calculate the upper threshold if there are values for the current medical name
    if not values_for_medical_name.empty:
        q1, q3 = np.percentile(values_for_medical_name, [25, 75])
        threshold_factor = 1.5
        iqr = q3 - q1
        upper_threshold = q3 + threshold_factor * iqr
        
        # convert medical_name to a string
        medical_name = str(medical_name)
        
        # create a new row for the output dataframe with the medical name and calculated values
        new_row = pd.DataFrame({'Medical Name': medical_name, 'Upper Threshold': upper_threshold}, index=[0])
    
        # append the new row to the output dataframe
        output_data = pd.concat([output_data, new_row], ignore_index=True)

# write the output dataframe to a csv file
output_data.to_csv('threshold_iqr.csv', index=False)
```

```python
import pandas as pd
from scipy.stats import iqr

# read the original CSV file
df = pd.read_csv('scores.csv')

# group the dataframe by the medical name column
grouped = df.groupby('Medicine Name')

# calculate the mean and IQR of each group
agg_dict = {'Value': ['mean', iqr]}
grouped_data = grouped.agg(agg_dict)

# flatten the column names
grouped_data.columns = ['_'.join(col).rstrip('_') for col in grouped_data.columns.values]

# write the means and IQRs to a new CSV file
grouped_data.to_csv('means_iqr_file.csv')
```
