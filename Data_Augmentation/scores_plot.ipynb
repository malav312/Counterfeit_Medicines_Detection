```python
import csv
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Create the "consistency_analysis" subdirectory if it doesn't already exist
if not os.path.exists("consistency_analysis"):
    os.makedirs("consistency_analysis")

# Open the CSV file
with open('scores.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)

    # Iterate over each row in the CSV file
    for row in reader:
        title = row[0]  # Get the title from the first column
        data = list(map(float, row[1:]))  # Get the data from the remaining columns and convert to float

        # Create the plot using Seaborn
        sns.set_style('darkgrid')
        fig, ax = plt.subplots()
        sns.lineplot(data=data, ax=ax)

        # Set the title and axis labels
        ax.set_title(title)
        ax.set_xlabel('Data Index')
        ax.set_ylabel('Metrics Recieved')

        # Save the plot as an image file in the "consistency_analysis" subdirectory with the title as the filename
        filename = f"consistency_analysis/{title}.png"
        fig.savefig(filename)

        # Close the plot to free up memory
        plt.close(fig)
```

```python
#Gaussian Plots
import csv
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Open the CSV file
with open('scores.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)

    # Skip the first row
    next(reader)

    # Iterate over each row in the CSV file
    for row in reader:
        title = row[0]  # Get the title from the first column
        data = np.array(list(map(float, row[1:])))  # Get the data from the remaining columns and convert to a numpy array

        # Fit a normal distribution to the data using scipy.stats
        mu, std = norm.fit(data)

        # Plot the histogram of the data with the fitted normal distribution overlaid
        plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
        xmin, xmax = plt.xlim()
        x = np.linspace(xmin, xmax, 100)
        p = norm.pdf(x, mu, std)
        plt.plot(x, p, 'k', linewidth=2)

        # Set the title and axis labels
        plt.title(title)
        plt.xlabel('Data Value')
        plt.ylabel('Frequency')

        # Save the plot as an image file with the title as the filename
        filename = f"guassian_analysis/{title}.png"
        plt.savefig(filename)

        # Close the plot to free up memory
        plt.close()
```
